```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
```

```python
est = []
act= []
f = open('estimations.csv')
for line in f:
    est += [eval(line)[0]]
    act += [eval(line)[1]]
```

```python
#est
```

```python
fig = plt.figure(figsize=(10,10))
plt.scatter([x for (x,y) in act], [y for (x,y) in act], color='blue', s=40, label='Gps-read')
plt.scatter([x for (x,y) in est], [y for (x,y) in est], color='red', alpha=.9, s=10, label='Estimator outputted')
plt.legend()
plt.plot()
```

```python
est_x = [x for (x,y) in est]
est_y = [y for (x,y) in est]
act_x = [x for (x,y) in act]
act_y = [y for (x,y) in act]
```

```python
fig, ax = plt.subplots(figsize=(10,10))

# draw initial plots
scat1 = ax.scatter(act_x[0], act_y[0], s=40, label='actuals', color='blue')
scat2 = ax.scatter(est_x[0], est_y[0], s=10, alpha=.9, label='estimated', color='red')
bounds = [[min(min(est_x), min(act_x)) - 2, max(max(est_x), max(act_x)) + 2], [min(min(est_y), min(act_y)) - 2, max(max(est_y), max(act_y)) + 2]]
ax.set(xlim=[bounds[0][0], bounds[0][1]], ylim=[bounds[1][0], bounds[1][1]], xlabel='x position', ylabel='y position')
ax.legend()


def update(frame):
    # for each frame, update the data stored on each artist.
    ax = act_x[:frame]
    ay = act_y[:frame]
    ex = est_x[:frame]
    ey = est_y[:frame]
    # update the actuals scatter plot
    data1 = np.stack([ax, ay]).T
    scat1.set_offsets(data1)
    # update the estimated scatter plot
    data2 = np.stack([ex, ey]).T
    scat2.set_offsets(data2)
    return (scat1, scat2)

# animation has 1 frame per each data point
ani = animation.FuncAnimation(fig=fig, func=update, frames=len(est_x), interval=30)
plt.show()
```

```python
# save to file
ani.save(filename="animation.html", writer="html")
```
